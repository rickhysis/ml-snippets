```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# experience from 1 to 8
experience = 10 * np.random.random((8, 1))

# salary = a + b * experience
salary = 125 * experience + 1000 + np.random.normal(size=experience.shape)

# create a linear regression model
model = LinearRegression()
model.fit(experience, salary)

# predict salary from the data
experience_new = np.linspace(1, 10, 8)
salary_new = model.predict(experience_new[:, np.newaxis])

# plot the results
plt.figure(figsize=(4, 3))
ax = plt.axes()
ax.scatter(experience, salary)
ax.plot(experience_new, salary_new)

ax.set_xlabel('experience')
ax.set_ylabel('salary')

ax.axis('tight')


plt.show()
```
